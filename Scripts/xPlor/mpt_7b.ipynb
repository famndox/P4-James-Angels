```python
import pandas as pd
import numpy as np
import re
```

```python
folk_df = pd.read_csv("../Resources/Datasets/folk_tales_deduplicated.csv", encoding='ISO-8859-1')

folk_df.head()
```

```python
column_values = folk_df.iloc[:, 3]
clean_values = column_values.apply(lambda x: re.sub(r'[^a-zA-Z0-9\s\.]', '', str(x)))
folk_df.iloc[:, 3] = clean_values
folk_df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(folk_df, test_size=0.1)

train.to_csv('dataset_train.csv', columns=["text"], header=True, index=False, sep=",")
test.to_csv('dataset_test.csv', columns=["text"], header=True, index=False, sep=",")
```
