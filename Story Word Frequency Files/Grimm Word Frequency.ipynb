```python
#Load the dataset
grimm_df=pd.read_excel('../Project 4 Resources/Datasets/grimm.xls')

#Display same data
grimm_df.head(25)
```

```python
#Drop NaN and leave blank
grimm_df.fillna('', inplace=True)

#Update columns
grimm_df = grimm_df.rename(columns={"Unnamed: 0": "Text"}).drop(columns=["Unnamed: 1"])

#Display removed rows
grimm_df.head(25)
```

```python
from collections import Counter

#Find most common value in the Title Text to see if there are any "Null" values
print(Counter(grimm_df['Text']).most_common(50))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

#Create a dictionary to house common words that appear is most folk tale, this list is arbitrary

word_freq = {}

words = ['evil', 'good', 'princess', 'prince', 'blood', 'afraid', 'fear', 'kill', 'son', 
    'daughter', 'power', 'luck', 'alive', 'magic', 'truth', 'huntsman', 'wife', 'husband', 'mother', 
    'father', 'children', 'white', 'devil', 'run', 'king', 'queen', 'witch', 'wisdom', 'travel',  
    'hurt', 'forest', 'castle', 'dragon', 'giant', 'peasant', 'enchanted',
    'apple', 'rose', 'raven', 'wolf', 'cottage', 'garden', 'treasure']

lemmatizer = WordNetLemmatizer()  # Create an instance of the WordNetLemmatizer class

for word in words:
    lemmatized_word = lemmatizer.lemmatize(word)
    word_freq[lemmatized_word] = word_freq.get(lemmatized_word, 0) + 1
```

```python
# Cycle through the "text" column and count the frequency of each word
for index, row in grimm_df.iterrows():
    text = row['Text']
    for word, freq in word_freq.items():
        if word in text.lower():  # Convert text to lowercase for case-insensitive search
            word_freq[word] += 1

# Create a pandas DataFrame from the word_freq dictionary
grimm_word_df = pd.DataFrame(list(word_freq.items()), columns=['word', 'frequency'])

# Print the word frequency to scan in the "text" column
print(grimm_word_df)
```

```python
#Export grimm word frequency count as own excel

grimm_word_df.to_excel('grimm_word_freq.xlsx', index=False)
```
