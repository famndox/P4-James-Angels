```python
import nltk
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

#Create a dictionary to house common words that appear is most folk tale, this list is arbitrary

word_freq = {}

words = ['evil', 'good', 'princess', 'prince', 'godmother', 'afraid', 'fear', 'kill', 'married', 'sisters', 
         'brothers', 'power', 'luck', 'wish', 'magic', 'truth', 'treasure', 'wife', 'husband', 'mother', 
         'father', 'children', 'sign', 'devil', 'penalty', 'king', 'queen', 'fairy', 'dragon', 'dream', 
         'wisdom', 'travel', 'journey', 'hurt']
lemmatizer = WordNetLemmatizer()  # Create an instance of the WordNetLemmatizer class

for word in words:
    lemmatized_word = lemmatizer.lemmatize(word)
    word_freq[lemmatized_word] = word_freq.get(lemmatized_word, 0) + 1
```

```python
# Cycle through the "text" column and count the frequency of each word
for index, row in folk_df.iterrows():
    text = row['text']
    for word, freq in word_freq.items():
        if word in text.lower():  # Convert text to lowercase for case-insensitive search
            word_freq[word] += 1

# Create a pandas DataFrame from the word_freq dictionary
folk_word_df = pd.DataFrame(list(word_freq.items()), columns=['word', 'frequency'])

# Print the word frequency to scan in the "text" column
print(folk_word_df)
```

```python
#Export word and frequency data as own excel

folk_word_df.to_excel('folk_word_freq.xlsx', index=False)
```

```python
#Count the frequency of the word with the "nation" column, this is interpretated as word to language

word_freq = {}

for index, row in folk_df.iterrows():
    text = row['text']
    nation = row['nation']  # Get the nation for this row
    for word in words:
        lemmatized_word = lemmatizer.lemmatize(word)
        if lemmatized_word in text.lower():  # Convert text to lowercase for case-insensitive search
            # Create a dictionary to store the word frequency for each nation
            nation_word_freq = word_freq.get(nation, {})
            nation_word_freq[lemmatized_word] = nation_word_freq.get(lemmatized_word, 0) + 1
            word_freq[nation] = nation_word_freq

print (word_freq)
```

```python
#Create a readable table for nation to word_freq

# Create a dataframe from the word_freq dictionary
df = pd.DataFrame(word_freq).T

# Set the index name as 'nations'
df.index.name = 'nations'

# Print the dataframe
print(df.round(1))
```

```python
#Fill NaN values with 'blank' as to not remove the entire row

df.fillna('', inplace=True)
```

```python
# Print the dataframe with the removed 'NaN' values
print(df.round(1))
```

```python
#Export nation word frequency count as own excel

df['nation'] = df.index
df.to_excel('nation_word_freq.xlsx', index=False)
```
